Função para enviar e-mail

```python
def envia_email(link_site, valor_real, produto):

    #criando conexão com o e-mail
    host = 'smtp.gmail.com'
    port = 587 #465
    user = 'jhonatandev0@gmail.com'
    password = 'jho240400'

    #login com servidor
    print('Realizando Login...')
    #server = smtplib.SMIP('smtp.gmail.com: 587')
    server = smtplib.SMTP(host, port)
    server.starttls()
    server.login(user, password)
    print("login realizado!")

    #cirando estrutura de email
    msg = email.message.Message()
    msg['Subject'] = f'Oportunidade de compra! O produto {produto} abaixou de preço!'
    msg['From'] = 'jhonatandev0@gmail.com'
    msg['To'] = 'jhonatandev0@gmail.com'
    conteudo_email = f'O preço do item abaixou! :D Hoje o seu preço está: R${valor_real}\n Acesse o link para realizar a compra: {link_site}'
    msg.set_payload(conteudo_email)

    #enviando mensagem
    print('Enviando mensagem...')
    server.sendmail(msg['From'], msg['To'], msg.as_string().encode('latin1'))
    print('Mensagem enviada!')
    server.quit()
```

```python
#Função que realiza registro no banco de dados
def inserir_registro(produto, valor_real, URL):
    string_conexao = gera_string_conexao()
    
    data_coleta = (datetime.datetime.utcnow())
    try:
        engine = sqlalchemy.create_engine(string_conexao)
        conn = engine.connect()
        print('Conexão realizada com sucesso')
        try:
            query = f"""insert into produto_coletado (nome_produto, valor_real, url)
                    values('{produto}',{valor_real},'{URL}')"""
            
            conn.execute(query)
            print(f'Registro inserido com sucesso: {produto}')
            
        except Exception as e:
            print(f"Não foi possível adicionar o item {produto}. Erro: {e}")
            
    except Exception as e:
        print(f'Erro de conexão: {e}')
```
